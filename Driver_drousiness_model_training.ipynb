```python
# importing modules 
import tensorflow as tf #for model making 
import cv2             #for geting camera and image detection
import matplotlib.pyplot as plt   #for ploting graph
import numpy as np                #for making data array
import os   #for path 
```

### reading and converting image into an array for data and labels 

```python
training_data=[]
Datadiractory="eyedata"
Classes=["closed_training_2","open_training_2"]
img_size=224
def create_training_data():
    for category in Classes:
        path=os.path.join(Datadiractory,category)
        class_num=Classes.index(category)
        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                backtorgb=cv2.cvtColor(img_array,cv2.COLOR_GRAY2RGB)
                new_img_array=cv2.resize(backtorgb,(img_size,img_size))
                training_data.append([new_img_array,class_num])
            except Exception as e:
                pass
```

```python
create_training_data()
```

```python
print(len(training_data))
```

```python
import random  #overfitting 

random.shuffle(training_data)
```

```python
#reshaping  and making input size same 
X=[]
y=[]

for features,label in training_data:
    X.append(features)
    y.append(label)
    
X=np.array(X).reshape(-1,img_size,img_size,3)
```

```python
X.shape
```

```python
#normalize the data 
X=X/255.0 
```

```python
Y=np.array(y)
```

```python
# saving data
import pickle

pickle_out=open("data/X.pickle","wb")
pickle.dump(X,pickle_out)
pickle_out.close()

pickle_out=open("data/y.pickle","wb")
pickle.dump(y,pickle_out)
pickle_out.close()
```

```python
#for skipping the above methods loading this data 
import pickle
pickle_in=open("data/X.pickle","rb")
X=pickle.load(pickle_in)

pickle_in=open("data/y.pickle","rb")
y=pickle.load(pickle_in)
```

# deep learning model -Transfer learning 


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# load old model
model=tf.keras.models.load_model('models/eye_open_close.h5')
# model.summary()
# new model
# model=tf.keras.applications.mobilenet.MobileNet()
```

## transfer learning

```python
base_input=model.layers[0].input  #input layer 

base_output=model.layers[-4].output #output layer 
```

```python
Flat_layer=layers.Flatten()(base_output)
final_output=layers.Dense(1)(Flat_layer)
final_output=layers.Activation('sigmoid')(final_output)
```

```python
new_model=keras.Model(inputs=base_input,outputs=final_output)
```

# settings for binary classification (open/closed)

```python
new_model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
new_model.fit(X,Y,epochs=3,validation_split=0.1) #training 
```
